```python
# Dados descritivos
df_mkt.describe()
```

```python
# Gráfico de corelação geral
plt.figure(figsize=(20,10))
sns.heatmap(df_mkt.corr(), annot=True)
```

```python
# Gráfico de Correlação de dispesão com "sales" por base em relação aos
# investimentos em "youtube", "facebook" e "newspaper"
plt.figure(figsize=(20,10))
sns.pairplot(df_mkt, x_vars=['youtube', 'facebook', 'newspaper'], y_vars='sales', height=4, aspect=1, kind='scatter')
```

## Etapa 02 - Análise Exploratória

```python
# Regressão lienar simples para avaliar o quanto:
# investir em "youtube" interfere no valor de "sales"
x_yt = df_mkt['youtube'].values.reshape(-1,1)
y_yt = df_mkt['sales']
x_train, x_test, y_train, y_test = train_test_split(x_yt, y_yt, test_size=0.2, random_state=42)
model_yt = LinearRegression()
model_yt.fit(x_train, y_train)
yt_score = model_yt.score(x_test, y_test)
```

```python
# Regressão lienar simples para avaliar o quanto:
# investir em "facebook" interfere no valor de "sales"
x_fb = df_mkt['facebook'].values.reshape(-1,1)
y_fb = df_mkt['sales']
x_train, x_test, y_train, y_test = train_test_split(x_fb, y_fb, test_size=0.2, random_state=42)
model_fb = LinearRegression()
model_fb.fit(x_train, y_train)
fb_score = model_fb.score(x_test, y_test)
```

```python
# Regressão lienar simples para avaliar o quanto:
# investir em "newspaper" interfere no valor de "sales"
x_nw = df_mkt['newspaper'].values.reshape(-1,1)
y_nw = df_mkt['sales']
x_train, x_test, y_train, y_test = train_test_split(x_nw, y_nw, test_size=0.2, random_state=42)
model_nw = LinearRegression()
model_nw.fit(x_train, y_train)
nw_score = model_nw.score(x_test, y_test)
```

```python
print(f'Score de "youtube": {yt_score:.4f}')
print(f'Score de "facebook": {fb_score:.4f}')
print(f'Score de "newspaper": {nw_score:.4f}')
```

Partindo do pressuposto de que o score (r²) é o quanto uma relação se aproxima podemos endenter que as campanhas do **Youtube** se correlaciona a **66,29%** sendo a mais relevante. Os valores investidos com **Facebook** demonstram ter como corelação **32,49%**, por quanto os investimentos em **Jornais** demonstra uma corelação muito baixa de **12,68%**.




## Etapa 03 - Modelagem

```python
x = df_mkt[['youtube', 'facebook', 'newspaper']]
y = df_mkt['sales']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Coeficientes = Influência de cada variável
model.coef_
```

```python
# Score R² = Porcentagem da variância no algo
model.score(x_test, y_test)
```

## Etapa 04 - Calcular Predição

```python
# predição do modelo
yPred = model.predict(x_test)
```

```python
# Graficuzinho
plt.scatter(y_test, yPred)
plt.xlabel("Valores reais")
plt.ylabel("Valores previstos")
plt.title("Valores reais vs. Valores previstos")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red') # linha diagonal
plt.show()
```
